```python
from textblob import TextBlob
import pandas as pd

df = pd.read_excel('data.xlsx')
df.set_index('No_of_Likes', inplace=True)

sentiment_results = []  # To store sentiment analysis results

for index, row in df.iterrows():
    y = row["text"]
    edu = TextBlob(y)
    x = edu.sentiment.polarity  # (-1, 0, 1)
    z = edu.sentiment.subjectivity

    if x < 0:
        result = f"{index} Negative {x} {z}"
    elif x == 0:
        result = f"{index} Neutral {x} {z}"
    elif x > 0:
        result = f"{index} Positive {x} {z}"

    sentiment_results.append(result)  # Save sentiment analysis results
    print(result)

# Add a new column 'Sentiment' to the DataFrame
df['Sentiment'] = sentiment_results

# Save the DataFrame to a new Excel file
output_file_path = 'sentiment_analysis_results.xlsx'
df.to_excel(output_file_path)

print(f"Sentiment analysis results saved to {output_file_path}")



```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import joblib
from textblob import TextBlob

# Read the data
df = pd.read_excel('data.xlsx')

# Perform sentiment analysis and create a 'Sentiment' column
df['Sentiment'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
df['Sentiment'] = df['Sentiment'].apply(lambda x: "Negative" if x < 0 else ("Neutral" if x == 0 else "Positive"))

# Encode the sentiment labels
label_encoder = LabelEncoder()
df['Sentiment'] = label_encoder.fit_transform(df['Sentiment'])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['Sentiment'], test_size=0.2, random_state=42)

# Vectorize the text data using TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Initialize classifiers
logistic_model = LogisticRegression()
svm_model = SVC(kernel='linear')
knn_model = KNeighborsClassifier()

# Train Logistic Regression model
logistic_model.fit(X_train_tfidf, y_train)

# Train SVM model
svm_model.fit(X_train_tfidf, y_train)

# Train KNN model
knn_model.fit(X_train_tfidf, y_train)

# Make predictions on the test set
logistic_predictions = logistic_model.predict(X_test_tfidf)
svm_predictions = svm_model.predict(X_test_tfidf)
knn_predictions = knn_model.predict(X_test_tfidf)

# Evaluate Logistic Regression model
logistic_accuracy = accuracy_score(y_test, logistic_predictions)
logistic_classification_report = classification_report(y_test, logistic_predictions)

# Evaluate SVM model
svm_accuracy = accuracy_score(y_test, svm_predictions)
svm_classification_report = classification_report(y_test, svm_predictions)

# Evaluate KNN model
knn_accuracy = accuracy_score(y_test, knn_predictions)
knn_classification_report = classification_report(y_test, knn_predictions)

print("Logistic Regression Model Accuracy:", logistic_accuracy)
print("Logistic Regression Classification Report:")
print(logistic_classification_report)

print("\nSVM Model Accuracy:", svm_accuracy)
print("SVM Classification Report:")
print(svm_classification_report)

print("\nKNN Model Accuracy:", knn_accuracy)
print("KNN Classification Report:")
print(knn_classification_report)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import joblib
from textblob import TextBlob

```

```python
# Read the data
df = pd.read_excel('data.xlsx')

# Perform sentiment analysis and create a 'Sentiment' column
def perform_sentiment_analysis(text):
    edu = TextBlob(text)
    polarity = edu.sentiment.polarity  # (-1, 0, 1)
    if polarity < 0:
        return "Negative"
    elif polarity == 0:
        return "Neutral"
    else:
        return "Positive"

# Parallelize sentiment analysis using joblib
df['Sentiment'] = joblib.Parallel(n_jobs=-1)(joblib.delayed(perform_sentiment_analysis)(text) for text in df['text'])

# Encode the sentiment labels
label_encoder = LabelEncoder()
df['Sentiment'] = label_encoder.fit_transform(df['Sentiment'])

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['Sentiment'], test_size=0.5, random_state=42, stratify=df['Sentiment'])

# Vectorize the text data using TF-IDF
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

```

```python
# Train a Logistic Regression model
model = LogisticRegression()
model.fit(X_train_tfidf, y_train)

# Make predictions on the test set
predictions = model.predict(X_test_tfidf)

# Evaluate the model
accuracy = accuracy_score(y_test, predictions)
classification_rep = classification_report(y_test, predictions, target_names=label_encoder.classes_, output_dict=True)

```

```python
print("Model Accuracy:", accuracy)
print("Classification Report:")
print(classification_rep)

# Save the model (optional)
model_filename = 'sentiment_model.joblib'
joblib.dump(model, model_filename)
print(f"Model saved to {model_filename}")

# User input for sentiment prediction
user_input = input("Enter a text for sentiment prediction: ")
user_input_tfidf = tfidf_vectorizer.transform([user_input])
user_prediction = model.predict(user_input_tfidf)

decoded_prediction = label_encoder.inverse_transform(user_prediction)
print(f"Predicted Sentiment: {decoded_prediction[0]}")
```
